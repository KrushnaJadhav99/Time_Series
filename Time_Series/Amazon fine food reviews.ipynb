```python
from nltk.corpus import stopwords

import string
string.punctuation
```

```python
def text_process(mess):            ### creating a function
    """                                                        ## a docstring
    1. remove the punctuation
    2. remove the stopwords
    3. return the list of clean textwords
    
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = "".join(nopunc)
    
    return [ word for word in nopunc.split() if word not in stopwords.words("english")]
```

```python
df['Text'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer  #ts used to get count each every word

bow_transformer= CountVectorizer(analyzer=text_process ).fit(df['Text'])

bow_transformer.vocabulary_
```

```python
len(bow_transformer.vocabulary_)
```

```python
Text_bow=bow_transformer.transform(df.Text)  ##creating tdm

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Text_bow,df.Score,test_size=0.2)

from sklearn.naive_bayes import MultinomialNB
naive_bay=MultinomialNB()
spam_nb_model=naive_bay.fit(x_train,y_train)

pred=naive_bay.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,pred)
```

```python
con_mat = confusion_matrix(pred,y_test)

print("This is Confusion Matrix:",)
print(con_mat)

fp=con_mat[1][0]
fn=con_mat[0][1]
tp=con_mat[1][1]
tn=con_mat[0][0]

Acc=(tp+tn)/(fp+fn+tp+tn)
FPR=(fp/(fp+tn))
TPR=(tp/(tp+fn))
Precision=(tp/(tp+fp))
Recall=TPR
f1_score=(2*Precision*Recall/(Precision+Recall))

df1=pd.DataFrame({'Naive Bayes':[round(Acc*100,2),round(FPR*100,2),round(TPR*100,2),round(Precision*100,2),round(f1_score*100,2)]},
index=(["Accuracy",'FPR','TPR/Recall','Precision','F1_score']))
df1
```
